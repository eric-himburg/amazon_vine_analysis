```python
# Import Dependencies
import pandas as pd 
import numpy as np
```

```python
# Make a reference to the vine_table.csv file path
csvpath = "Resources/vine_table.csv"
# Import the books.csv file as a DataFrame with utf-8 encoding.
vine_review_df=pd.read_csv(csvpath,encoding="utf-8")
vine_review_df.head()
```

```python
review_gr20_df=vine_review_df.loc[vine_review_df['total_votes']>=20]
review_gr20_df.head()
```

```python
helpful_review_df=review_gr20_df.loc[(review_gr20_df['helpful_votes']/review_gr20_df['total_votes'])>=0.5]
helpful_review_df.head()
```

```python
vine_unpaid_df=helpful_review_df[helpful_review_df.vine=='N']
vine_unpaid_df.head()
```

```python
helpful_review_df["vine"].count()
helpful_review_df.loc[helpful_review_df["vine"]=='Y'].count()
```

```python
vine_paid_df=vine_review_df.loc[vine_review_df["vine"]=='Y']
vine_paid_df.head()
```

```python
# create a 5-star review summary list for vine reviews
total_reviews=vine_paid_df["review_id"].count()
fivestar_reviews=np.sum(vine_paid_df['star_rating'] == 5.0)
percent_fivestar=round((fivestar_reviews/total_reviews)*100,1)
vine_list=("vine",total_reviews,fivestar_reviews,percent_fivestar)
vine_list
```

```python
# create a 5-star review summary list for non-vine reviews
total_reviews=vine_unpaid_df["review_id"].count()
fivestar_reviews=np.sum(vine_unpaid_df['star_rating'] == 5.0)
percent_fivestar=round((fivestar_reviews/total_reviews)*100,1)
nonvine_list=("non-vine",total_reviews,fivestar_reviews,percent_fivestar)
nonvine_list
```

```python
# create a summary database which compares vine to non-vine reviews
column_names=["review type","total reviews","total 5-star reviews","% 5-star reviews"]
summary_df=pd.DataFrame(columns=column_names)
summary_df.loc[0]=vine_list
summary_df.loc[1]=nonvine_list
summary_df.set_index('review type', inplace=True)
summary_df
```
